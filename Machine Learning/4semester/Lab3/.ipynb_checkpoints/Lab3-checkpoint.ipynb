```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt

from sklearn.svm import SVC
from matplotlib.colors import ListedColormap

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

1.1 Transforming data into correct format

```python
df = pd.read_csv(os.path.abspath(os.path.join(os.getcwd(), '1895-2023.csv')), skiprows=4)
df.Date = df.Date // 100
df = df.rename(columns={'Value': 'Temperature'})
df.head()
```

1.2 splitting data into training and testing sets

```python
x_train, x_test, y_train, y_test = train_test_split(df.Date.values.reshape(-1, 1), df.Temperature.values,
                                                    random_state=11)
```

1.3 training the model

```python
linear_regression = LinearRegression()
linear_regression.fit(x_train, y_train)
print(linear_regression.coef_)
print(linear_regression.intercept_)
```

1.4 checking the results

```python
prediction = linear_regression.predict(x_test)
expected = y_test

for p, e in zip(prediction, expected):
    print(f'predicted: {p: .2f} expected: {e: .2f}')
```

1.5 Lets use obtained data about model for prediction of future data

```python
def predict(x):
    return linear_regression.coef_ * x + linear_regression.intercept_


predict(2024)
predict(1890)
```

1.6 Lets visualize our data and regression line

```python
#data
axes = sns.scatterplot(data=df, x='Date', y='Temperature', hue='Temperature', palette='autumn')
axes.set_ylim(10, 70)

#regression line
x = np.array([min(df.Date), max(df.Date)])
y = predict(x)
line = plt.plot(x, y, color='red')
```

2.1 Extract code from the example from lecture 7, generate a data set and
classify it using the SVC classifier (slide 95).

```python

np.random.seed(1)
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)
plt.scatter(X_xor[y_xor == 1, 0], X_xor[y_xor == 1, 1],
            c='b', marker='x', label='1')
plt.scatter(X_xor[y_xor == -1, 0], X_xor[y_xor == -1, 1],
            c='r', marker='s', label='-1')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.legend(loc='best')
plt.show()


def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    markers = ('s', 'x', 'o', 1, 1)
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))

    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    for idx, c1 in enumerate(np.unique(y)):
        plt.scatter(x=X[y == c1, 0],
                    y=X[y == c1, 1],
                    alpha=0.8,
                    c=colors[idx],
                    marker=markers[idx],
                    label=c1)

    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100,
                    label='Zestaw testowy')


svm = SVC(kernel='rbf', random_state=1, gamma=0.10, C=10.0)
svm.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.legend(loc='upper left')
plt.show()
```

3. Let`s compare several classification estimators
such as KNeighborsClassifier, SVC, and GaussianNB for the same dataset built into scikit-learn (choose any one you want). For out purposes I will use dataset of labeled images, which is often used to evaluate the efficiency of a given classifier

```python
# Load the digits dataset
digits = load_digits()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3, random_state=42)

# Create instances of the classifiers
knn = KNeighborsClassifier()
svm = SVC()
gnb = GaussianNB()

# Fit the classifiers to the training data
knn.fit(X_train, y_train)
svm.fit(X_train, y_train)
gnb.fit(X_train, y_train)

# Make predictions on the testing data
knn_pred = knn.predict(X_test)
svm_pred = svm.predict(X_test)
gnb_pred = gnb.predict(X_test)

# Evaluate the performance of each classifier
#accuracy = correct/total
knn_acc = accuracy_score(y_test, knn_pred)
svm_acc = accuracy_score(y_test, svm_pred)
gnb_acc = accuracy_score(y_test, gnb_pred)

knn_cm = confusion_matrix(y_test, knn_pred)
svm_cm = confusion_matrix(y_test, svm_pred)
gnb_cm = confusion_matrix(y_test, gnb_pred)

#precision = TruePositive / (TruePositive + FalsePositive)
knn_precision = precision_score(y_test, knn_pred, average='weighted')
svm_precision = precision_score(y_test, svm_pred, average='weighted')
gnb_precision = precision_score(y_test, gnb_pred, average='weighted')

# Recall, also known as sensitivity, is a performance metric that measures
# the ability of a classifier to correctly identify all positive samples in a dataset.
# It is calculated as the ratio of true positives to the sum of true positives and false negatives:
# Recall = True Positives / (True Positives + False Negatives)
knn_recall = recall_score(y_test, knn_pred, average='weighted')
svm_recall = recall_score(y_test, svm_pred, average='weighted')
gnb_recall = recall_score(y_test, gnb_pred, average='weighted')

#combines precision and recall scores
knn_f1 = f1_score(y_test, knn_pred, average='weighted')
svm_f1 = f1_score(y_test, svm_pred, average='weighted')
gnb_f1 = f1_score(y_test, gnb_pred, average='weighted')

# Print the results
print("KNN accuracy:", knn_acc)
print("SVM accuracy:", svm_acc)
print("GaussianNB accuracy:", gnb_acc)

print("KNN confusion matrix:\n", knn_cm)
print("SVM confusion matrix:\n", svm_cm)
print("GaussianNB confusion matrix:\n", gnb_cm)

print("KNN precision:", knn_precision)
print("SVM precision:", svm_precision)
print("GaussianNB precision:", gnb_precision)

print("KNN recall:", knn_recall)
print("SVM recall:", svm_recall)
print("GaussianNB recall:", gnb_recall)

print("KNN f1 score:", knn_f1)
print("SVM precision:", svm_f1)
print("GaussianNB precision:", gnb_f1)
```
