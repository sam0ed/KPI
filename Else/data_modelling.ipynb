```python
import numpy as np
import matplotlib.pyplot as plt

#   
n = 100

#  
time = np.linspace(0, 10, n)

#    
true_values = time**2

#   
measurement_error = np.random.normal(0, 5, n)

#    
#      
anomaly_probability = 0.1
anomaly_amplitude = 50
anomalies = np.random.uniform(-anomaly_amplitude, anomaly_amplitude, n)
anomaly_indices = np.random.uniform(0, 1, n) < anomaly_probability

#   ,     
measured_values = true_values + measurement_error
measured_values[anomaly_indices] += anomalies[anomaly_indices]

#  
plt.figure(figsize=(10, 6))
plt.plot(time, true_values, label="True Values (Quadratic)")
plt.scatter(time, measured_values, label="Measured Values", color="r")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Quadratic Measurement Model with Normal Error and Uniform Anomalies")
plt.legend()
plt.show()
```
