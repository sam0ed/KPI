```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import  confusion_matrix
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

import warnings
```

1.1 Let's visualize numbers coded in load_digits.imagees using matplotlib

```python
digits = load_digits()
```

1.2 Visualization of first 24 numbers

```python
figure, axes = plt.subplots(nrows=4, ncols=6, figsize=(6, 4))
for item in zip (axes.ravel(), digits.images, digits.target):
    axes, image, target = item
    axes.imshow(image, cmap=plt.cm.gray_r)
    axes.set_xticks([])
    axes.set_yticks([])
    axes.set_title(target)
plt.tight_layout()
```

1.3 Visualization of first 36 numbers

```python
figure, axes = plt.subplots(nrows=6, ncols=6, figsize=(6, 6))
for item in zip (axes.ravel(), digits.images, digits.target):
    axes, image, target = item
    axes.imshow(image, cmap=plt.cm.gray_r)
    axes.set_xticks([])
    axes.set_yticks([])
    axes.set_title(target)
plt.tight_layout()
```

2.1 Let's split the data into training and testing sets. Let's make the testing set of the size equal to 20 percent of the initial data, which is different from default splitting.

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, random_state=42, test_size=0.2)
```

3.1 Let's create the model and train it on the split data.

```python
knn= KNeighborsClassifier()
knn.fit(X=X_train, y= y_train)
```

4.1 Let's predict classes for the test data.

```python
predicted = knn.predict(X_test)
```

5.1 Let's compare obtained prediction with actual labels of the data.

```python
indices = [20, 24, 36]
for index in indices:
    if index < len(predicted) and index < len(y_test):
        print('Predicted values for predicted[:{index}] are:')
        print (predicted[:index])
        print('Actual labels of the data look like:')
        print (y_test[:index])
    else:
        print("Index out of range.")
```

6. Let's understand the performance of the model a bit better using different evaluation metrics for the model.

6.1 To get general understanding on how well model performs on the test data let's use built in score method of our model. As we can see the results are impressive.

```python
print (f'{knn.score(X_test, y_test):.3%}')
```

6.2 The second method to use is confusion matrix of the classifier. The rows represent the true labels of the sample, the column represents the label which was given to the sample during classification process.

```python
confusion = confusion_matrix(y_test, predicted)
confusion
```

7.1 Another useful metric is classification report which gives such important measures as precision recall etc.

```python
print(classification_report(y_test, predicted))
```

8.1 Let's use support vector machines and naive bayes to classify the digits and the compare the performance of the models.

```python
svc_classifier = SVC()
svc_classifier.fit(X_train, y_train)
svc_predictions = svc_classifier.predict(X_test)

gnb_classifier = GaussianNB()
gnb_classifier.fit(X_train, y_train)
gnb_predictions = gnb_classifier.predict(X_test)

print('SVC classifier score is: ', f'{svc_classifier.score(X_test, y_test):.3%}')
print('GNB classifier score is: ', f'{gnb_classifier.score(X_test, y_test):.3%}')
```

In general the GNB performance is worse than performance of SVC and KNN.

9.1 Let's test the performance of KNN by tuning the hyperparameter value from 1 to amount of samples present in the training data.

```python
warnings.filterwarnings("ignore", category=FutureWarning)

for k in range(1, len(X_train), 100):
    knn= KNeighborsClassifier(k)
    knn.fit(X=X_train, y= y_train)
    print(f'The performance of the model for value of k = {k} is: ', f'{knn.score(X_test, y_test):.3%}')

warnings.filterwarnings("default", category=FutureWarning)
```
