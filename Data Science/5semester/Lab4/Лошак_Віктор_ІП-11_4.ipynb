```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import datetime
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome()

url = "https://rozetka.com.ua/ua/mobile-phones/c80003/producer=apple/#search_text=iphone"
driver.get(url)
page_content = driver.page_source

driver.quit()
```

```python
def extract_product_details(soup):
    product_details = []
    product_containers = soup.find_all('div', class_='goods-tile__inner')

    for product in product_containers:

        name = product.find('span', class_='goods-tile__title').get_text(strip=True)
        price = product.find('span', class_='goods-tile__price-value').get_text(strip=True)
        availability = product.find('div', class_='goods-tile__availability').get_text(strip=True)

        product_details.append({'name': name, 'price': price, 'availability': availability})

    return product_details

```

```python
# run this cell to create a timestamped csv file with the results
now = datetime.datetime.now()
soup = BeautifulSoup(page_content, 'html.parser')
product_details = pd.DataFrame(extract_product_details(soup))
product_details.to_csv(f'rozetka_{now.date()}.csv', index=False)
```
