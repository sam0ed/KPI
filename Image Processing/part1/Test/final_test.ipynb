```python
from skimage import io, img_as_float
from skimage.metrics import structural_similarity as ssim
import cv2
import matplotlib.pyplot as plt
```

```python

def load_and_preprocess_image(path, slot = None):
    img = cv2.imread(path)
    img = cv2.resize(img, (888, 496)) 
    
    if slot is not None:
        slot.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        slot.axis('off') 
    return img

# Specify the paths to the two images directly
image_path1 = 'assets\dolorian1.png'
image_path2 = 'assets\dolorian2.png'


fig, axs = plt.subplots(1, 2, figsize=(10, 5))

image1 = load_and_preprocess_image(image_path1, axs[0])
image2 = load_and_preprocess_image(image_path2, axs[1])
plt.show()
```

```python
image1.shape, image2.shape
```

```python

similarity = ssim(image1, image2, multichannel=True, channel_axis=-1)
print(f"SSIM between the two images: {similarity}")
```
